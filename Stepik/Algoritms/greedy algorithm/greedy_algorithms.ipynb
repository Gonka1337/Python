# Задача 1.Покрытие точек отрезками

Вход: множество n точек на прямой x1,....,xn 

Выход: минимальное количество отрезков единичной длины, которыми можно покрыть все точки

## Надёжный шаг

Существует оптимальное покрытие, в котором самая левая точка покрыта левым концом отрезка.

Поэтому можно сразу добавить в решение отрезок, левый конец которого совпадает с этой минимальной точкой

## Алгоритм

```python
s=[1,2,3,4,5,6,7]
a=[]
while len(s)!=0:
    a.append([min(s),min(s)+1])
    for i in range(min(s),min(s)+2):
        if i in s:
            s.remove(i)
        
print(a)
```

## Улучшенный Алгоритм

```python
d=[1,2,3,4,5,6,7,8,9]
a=[]
d=sorted(d)
i=0
while i<=len(d)-1:
    [l,r]=d[i],d[i]+1
    a.append([l,r])
    i+=1
    while i<=len(d)-1 and d[i]<=r:
        i+=1
print(a)
```
