# покрыть отрезки точками

```python
a = [[int(i) for i in input().split()] for i in range(int(input()))]
a.sort(key=lambda i:i[1])
s=[]
for i in a:
    if s==[]:
        s.append(i[1])
    else:
        if i[0]>s[-1]:
            s.append(i[1])
        else:
            continue
print(len(s))
for i in s:
    print(i,end=' ')

        
            
        
        
        
        
```

# Задача на программирование: непрерывный рюкзак

```python
a=[int(i) for i in input().split()]
n,w=a[0],a[1]
d=0
s=[[int(i) for i in input().split()] for i in range(n)]
s.sort(key=lambda i:i[0]/i[1],reverse = True)
for i in s:
    if w!=0 and w>0:
        if w>=i[1]:
            d+=i[0]
            w-=i[1]
        else:
            d+=(i[1]+w-i[1])*(i[0]/i[1])
            w-=i[1]
    else:
        break
print('%.3f'%d)
```

# Задача на программирование: различные слагаемые

```python
n=int(input())
c=0
s=[]
for i in range(1,n+1):
    if c==n:
        break
    else:
        c+=i
        s.append(i)
        if sum(s)==n:
            break
print(len(s))
for i in s:
    print(i,end=' ')
            
        
    
```

```python
n=int(input())
c=1
s=[]
while c!=n:
    for i in range(1,n+1):
        if c+i==n:
            s.append([c,i])
            break
        else:
            c+=1
```
