# Установка дополнительных модулей

```python
import requests
```

```python
r = requests.get('https://www.twitch.tv/buster') # простой GET запрос
print(r.text) #вывод ответа от сервера
```

```python
url='https://www.twitch.tv/buster'
par ={'key1':'value1', 'key2':'value2'}
r = requests.get(url,params=par) #Передача параметров в запрос
print(r.url) # Сформированный url-адрес с учетом параметров GET запроса
```

```python
url='https://www.twitch.tv/buster'
cookies ={'cookies_are':'working'}
r = requests.get(url,cookies=cookies) #отправка сформированных cookies на сервер
print(r.text)
```

```python
print(r.cookies['example_cookie_name']) # использование cookies, полученных от сервера
```

Скачайте файл. В нём указан адрес другого файла, который нужно скачать с использованием модуля requests и посчитать число строк в нём.

Используйте функцию get для получения файла (имеет смысл вызвать метод strip к передаваемому параметру, чтобы убрать пробельные символы по краям).

После получения файла вы можете проверить результат, обратившись к полю text. Если результат работы скрипта не принимается, проверьте поле url на правильность. Для подсчёта количества строк разбейте текст с помощью метода splitlines.

В поле ответа введите одно число или отправьте файл, содержащий одно число.

```python
with open ('dataset_3378_2 (1).txt','r') as inf:
    c=0
    for line in inf:
        line=line.strip()
        r=requests.get(line)
        d=r.text
        for i in d.splitlines():
            c+=1
        print(c)
```

Имеется набор файлов, каждый из которых, кроме последнего, содержит имя следующего файла.

Скачайте предложенный файл. В нём содержится ссылка на первый файл из этого набора.

Все файлы располагаются в каталоге по адресу:
https://stepic.org/media/attachments/course67/3.6.3/

Загрузите содержимое ﻿последнего файла из набора, как ответ на это задание.

```python
with open ('dataset_3378_3.txt','r') as inf:
    for line in inf:
        line=line.strip()
        r=requests.get(line)
        for line in r.text.splitlines():
            r=requests.get('https://stepic.org/media/attachments/course67/3.6.3/'+line)
            print(r.url)
            continue
```

```python
with open ('dataset_3378_3 (1).txt','r') as inf:
    for line in inf:
        line=line.strip()
        r=requests.get(line)
        for line in r.text.splitlines():
            while line in r.text.splitlines():
                r=requests.get('https://stepic.org/media/attachments/course67/3.6.3/'+line)
                line=r.text
                print(r.text)
            else:
                continue
```
